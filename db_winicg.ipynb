```sql
/****** Object:  StoredProcedure [ISO].[SP_ADD_RISK]    Script Date: 4/3/2024 8:42:49 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- =============================================
-- Author:		<Author,,Name>
-- Create date: <Create Date,,>
-- Description:	<Description,,>
-- =============================================
CREATE PROCEDURE [ISO].[SP_ADD_RISK]
	-- Add the parameters for the stored procedure here
	@IDRULE VARCHAR(15) NULL,
	@NAMERISKS VARCHAR(50)NULL,
	@ORIGEN VARCHAR(50) NULL, 
	@STATE BIT NULL, 
	@CONSEQUENSE VARCHAR(500) NULL, 
	@SOURCE_RISK VARCHAR(100) NULL
AS
BEGIN
	DECLARE @COUNTER_ID INT,
	@IDRISK VARCHAR(15),
		@LASTCODE VARCHAR(15),
		@COUNTER INT,
		@SPLIT_COUNTER INT = 0


	SET @COUNTER_ID = (SELECT COUNT(*) FROM ISO.RISKS)

	IF(@COUNTER_ID<10)
	BEGIN
	--OBTENEMOS EL CODIGO MINIMO DE LA TABLA
           SET @LASTCODE = (SELECT RIGHT(MAX(IDRISKS),1) FROM ISO.RISKS) 
		   SET @SPLIT_COUNTER = 1
		 END
            ELSE IF (@COUNTER <100) 
			BEGIN
                SET @LASTCODE = (SELECT RIGHT(MAX(IDRISKS),2) FROM ISO.RISKS) 
				SET @SPLIT_COUNTER = 2
			END
                ELSE IF(@COUNTER <1000)
				BEGIN
                    SET @LASTCODE = (SELECT RIGHT(MAX(IDRISKS),3) FROM ISO.RISKS) 
					SET @SPLIT_COUNTER = 3
                END

SET @COUNTER = (SELECT COUNT(*)+1 FROM ISO.RISKS)
	
	SET @IDRISK = ISO.[GET_CODE_RISKS](@COUNTER)

	--COMPARAMOS EL ULTIMO DIGITO DE LA TABLA 
	IF @LASTCODE = RIGHT(@IDRISK,@SPLIT_COUNTER)
	BEGIN 
	SET @LASTCODE = @LASTCODE + 1
	SET @COUNTER = @LASTCODE
	SET @IDRISK = ISO.[GET_CODE_RISKS](@COUNTER)
	END
	-- SET NOCOUNT ON added to prevent extra result sets from
	-- interfering with SELECT statements.
	SET NOCOUNT ON;

    INSERT INTO ISO.RISKS(IDRISKS, IDRULE,NAMERISKS,ORIGEN,CONSEQUENSE,SOURCE_RISK,STATE, CREATEDATE, UPDATEDATE)
	VALUES (@IDRISK,@IDRULE, @NAMERISKS,@ORIGEN,@CONSEQUENSE, @SOURCE_RISK, 1, GETDATE(), NULL)

	SELECT * FROM ISO.RISKS R, ISO.ISORULE IR, AUD.FILES F WHERE R.IDRULE = IR.IDRULE AND R.IDRISKS =  @IDRISK
END
GO
```

```sql
/****** Object:  StoredProcedure [ORG].[SP_ADD_COMPANY_POSITION]    Script Date: 4/3/2024 8:42:49 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- =============================================
-- Author:		<Author,,Name>
-- Create date: <Create Date,,>
-- Description:	<Description,,>
-- =============================================
CREATE PROCEDURE [ORG].[SP_ADD_COMPANY_POSITION]
	-- Add the parameters for the stored procedure here
	@IDUSER VARCHAR(15) NULL, 
	@IDPROCESS VARCHAR(15) NULL, 
	@MANDATED VARCHAR(50) NULL, 
	@DESCRIPTION VARCHAR(1500) NULL,
	@RESPONSABILITIES VARCHAR(50) NULL, 
	@PROFILE_POSITION VARCHAR(50) NULL

AS
BEGIN
	DECLARE @COUNTER_ID INT,
	@ID_COMPANYPOSITION  VARCHAR(15),
		@LASTCODE VARCHAR(15),
		@COUNTER INT,
		@SPLIT_COUNTER INT = 0


	SET @COUNTER_ID = (SELECT COUNT(*) FROM ORG.COMPANY_POSITION)

	IF(@COUNTER_ID<10)
	BEGIN
	--OBTENEMOS EL CODIGO MINIMO DE LA TABLA
           SET @LASTCODE = (SELECT RIGHT(MAX(IDCOMPANY_POSITION),1) FROM ORG.COMPANY_POSITION) 
		   SET @SPLIT_COUNTER = 1
		 END
            ELSE IF (@COUNTER <100) 
			BEGIN
                SET @LASTCODE = (SELECT RIGHT(MAX(IDCOMPANY_POSITION),2) FROM ORG.COMPANY_POSITION) 
				SET @SPLIT_COUNTER = 2
			END
                ELSE IF(@COUNTER <1000)
				BEGIN
                    SET @LASTCODE = (SELECT RIGHT(MAX(IDCOMPANY_POSITION),3) FROM ORG.COMPANY_POSITION) 
					SET @SPLIT_COUNTER = 3
                END

SET @COUNTER = (SELECT COUNT(*)+1 FROM ORG.COMPANY_POSITION)
	
	SET @ID_COMPANYPOSITION = ORG.[GET_CODE_COMPANY](@COUNTER)

	--COMPARAMOS EL ULTIMO DIGITO DE LA TABLA 
	IF @LASTCODE = RIGHT(@ID_COMPANYPOSITION,@SPLIT_COUNTER)
	BEGIN 
	SET @LASTCODE = @LASTCODE + 1
	SET @COUNTER = @LASTCODE
	SET @ID_COMPANYPOSITION = ORG.[GET_CODE_COMPANY](@COUNTER)
	END
	-- SET NOCOUNT ON added to prevent extra result sets from
	-- interfering with SELECT statements.
	SET NOCOUNT ON;
	INSERT INTO ORG.COMPANY_POSITION (IDCOMPANY_POSITION, IDUSER, IDPROCESS, MANDATED, DESCRIPTION, RESPONSABILITIES, PROFILE_POSITION, CREATEDATE, UPDATEDATE)
	VALUES (@ID_COMPANYPOSITION, @IDUSER, @IDPROCESS, @MANDATED, @DESCRIPTION, @RESPONSABILITIES, @PROFILE_POSITION, GETDATE(), NULL )
    -- Insert statements for procedure here
	SELECT * FROM ORG.COMPANY_POSITION CP, US.USERS U, PRO.PROCESS P 
	WHERE CP.IDPROCESS =  P.IDPROCESS AND CP.IDUSER =  U.IDUSER AND 
	CP.IDCOMPANY_POSITION =  @ID_COMPANYPOSITION
END
GO
```
